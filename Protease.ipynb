```python
import pandas as pd
import numpy as np
import csv
```

```python
f = open('../PR.txt', 'r')
flag = False
no_drug_lst = []
drug_lst = []
for line in f:
    if flag == False:
        flag = True
    else:
        tokens = line.split("\t")
        if tokens[-1].strip():
            ptid = int(tokens[1])
            drug = tokens[6]
            seq = tokens[-1].strip()
            if drug == 'None':
                no_drug_lst.append([ptid, drug, seq])
            else:
                drug_lst.append([ptid, drug, seq])
```

```python
# sorted_ptid = sorted(arr, key=lambda x: int(x[0]))
```

```python
no_drug_df = pd.DataFrame(no_drug_lst, columns=['ptid', 'Drug', 'seq'])
drug_df = pd.DataFrame(drug_lst, columns=['ptid', 'Drug', 'seq'])

merged = pd.merge(no_drug_df, drug_df, on='ptid', how='inner', suffixes=('_ND', '_D'))
```

```python
merged.to_csv('alldata.csv', sep="\t", index=False, quoting=csv.QUOTE_NONE)
```

```python
# Encode + Remove noise
def encoding(seq):
    temp = ''
    for each_char in seq:
        encoded_char = encode_char(each_char.lower())
        if encoded_char:
            temp += encoded_char
        else: #encode_char() return None
            return None
    return temp

def encode_char(x):
    if x=='a':
        return '1'
    elif x=='t':
        return '2'
    elif x=='c':
        return '3'
    elif x=='g':
        return '4'
    else:
        print '%s invalid ' % (x)
        return None
    

encoded = []
for row in merged.itertuples(index=False):
    seq_none_enc = encoding(row[2].strip())
    seq_drug_enc = encoding(row[4].strip())
    if seq_none_enc and seq_drug_enc:
        encoded.append([seq_drug_enc, seq_none_enc])
    
```

```python
MAXLENGTH = len(max(encoded, key=len)[0])
# Create nodes
from pybrain.structure import LinearLayer, SigmoidLayer
inLayer = LinearLayer(MAXLENGTH)
hiddenLayer = SigmoidLayer(3)
outLayer = LinearLayer(MAXLENGTH)
```

```python
# Add nodes to network
n.addInputModule(inLayer)
n.addModule(hiddenLayer)
n.addOutputModule(outLayer)
```

```python
# Create Weight 
from pybrain.structure import FullConnection
in_to_hidden = FullConnection(inLayer, hiddenLayer)
hidden_to_out = FullConnection(hiddenLayer, outLayer)
```

```python
n.addConnection(in_to_hidden)
n.addConnection(hidden_to_out)
```
