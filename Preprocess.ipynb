```python
clean_seq = lambda x: x.strip().strip('.')

f = open('../ACTG302_RT_fasta.txt', 'r')
ptid = None
sequences = []
while True:
    desc = f.readline()
    seq = f.readline()
    if not seq: 
        break  # EOF
    tokens = desc.split("|")
#     print tokens
    ptid_new = tokens[0].split()[1]
    if ptid_new != ptid:
        if ptid is not None:
            sequences.append([start_seq, end_seq]) # Write previous record
        start_seq = clean_seq(seq)
        end_seq = None
        ptid = ptid_new
    else:
        end_seq = clean_seq(seq)
```

```python
count = 0
for pt in sequences:
    if pt[1] is None:
        count += 1
        sequences.remove(pt)
count
```
